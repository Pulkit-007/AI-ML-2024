```python
## Index columns and training delimiters
```

```python
data0 = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data0),index_col=0)
```

```python
pd.read_csv(StringIO(data0),index_col=2)
```

```python
data00 = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data00))
```

```python
pd.read_csv(StringIO(data00),index_col=False)
```

```python
## Combining usecols and index_col
data000 = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data000), usecols=['b', 'c'],index_col=False)
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data4 = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data4),escapechar='\\')
```
